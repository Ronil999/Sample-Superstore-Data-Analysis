```python
# Convert the order date column to a datetime object
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Group the data by order date and sum the sales
sales_by_date = df.groupby('Order Date')['Sales'].sum()

# Create a line chart to visualize the sales over time
plt.plot(sales_by_date)
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.title('Sales Over Time')
plt.show()
```
